## Analysis

```python
# Home team record vs FTR
tmp_df = df.filter(['home_prev_3_win','FTR'])
tmp_df = tmp_df.groupby('FTR').mean().reset_index()
tmp_df
plt.bar(x=tmp_df['FTR'], height=tmp_df['home_prev_3_win'])
plt.title('Average Home Team Wins in Last 3 Matches')
plt.ylabel("Average wins")
plt.xlabel("Results")
```

```python
# Home team goals scored vs FTR
tmp_df = df.filter(['home_prev_3_gs','FTR'])
tmp_df = tmp_df.groupby('FTR').mean().reset_index()
tmp_df
plt.bar(x=tmp_df['FTR'], height=tmp_df['home_prev_3_gs'])
plt.title('Average Home Team Goals Scored in Last 3 Matches')
plt.ylabel("Average goals scored")
plt.xlabel("Results")
```
