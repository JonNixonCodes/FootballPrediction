## Feature Engineering

```python
# Feature table filters
valid_seasons = ['0809','0910','1011','1112','1213','1314','1415','1516','1617','1718','1819'] # Valid seasons
valid_divs = ['E0'] # Valid division
keep_cols = ['HomeTeam','AwayTeam','Date','Season','Div','FTR'] # Keep columns
# Create feature table
feature_df = df.copy()
feature_df = feature_df[(feature_df['Season'].isin(valid_seasons)) & (feature_df['Div'].isin(valid_divs))]
feature_df = feature_df.filter(keep_cols)
# Add feature variables
feature_df['home_prev_8_draw'] = feature_df.apply(lambda x: ft.prev_n_draw(df,team=x['HomeTeam'],n=8,date=x['Date'],homeaway='both'),axis=1)
feature_df['home_prev_8_gc'] = feature_df.apply(lambda x: ft.prev_n_gc(df,team=x['HomeTeam'],n=8,date=x['Date'],homeaway='both'),axis=1)
feature_df['home_prev_8_gs'] = feature_df.apply(lambda x: ft.prev_n_gs(df,team=x['HomeTeam'],n=8,date=x['Date'],homeaway='both'),axis=1)
feature_df['away_prev_8_draw'] = feature_df.apply(lambda x: ft.prev_n_draw(df,team=x['AwayTeam'],n=8,date=x['Date'],homeaway='both'),axis=1)
feature_df['away_prev_8_gc'] = feature_df.apply(lambda x: ft.prev_n_gc(df,team=x['AwayTeam'],n=8,date=x['Date'],homeaway='both'),axis=1)
feature_df['away_prev_8_gs'] = feature_df.apply(lambda x: ft.prev_n_gs(df,team=x['AwayTeam'],n=8,date=x['Date'],homeaway='both'),axis=1)
feature_df['prev_8_most_wins'] = feature_df.apply(lambda x: ft.prev_n_most_wins(df,homeTeam=x['HomeTeam'],awayTeam=x['AwayTeam'],n=8,date=x['Date'],homeaway='both'),axis=1)
feature_df['prev_8_matchup_wins'] = feature_df.apply(lambda x: ft.prev_n_matchup_wins(df,homeTeam=x['HomeTeam'],awayTeam=x['AwayTeam'],n=8,date=x['Date'],homeaway='both'),axis=1)
feature_df # Show
```

```python
# Import libraries
from sklearn.preprocessing import OneHotEncoder
```

```python
cat_feature = feature_df.dtypes[feature_df.dtypes == object].keys()
cat_feature = cat_feature.drop(['HomeTeam','AwayTeam','Season','Div','FTR'])
cat_feature
```

```python
# Prepare data for training
cat_feature = feature_df.dtypes[feature_df.dtypes == object].keys()
cat_feature = cat_feature.drop(['HomeTeam','AwayTeam','Season','Div','FTR'])
# Encode categorical features using One-Hot-Encoding
encoder = OneHotEncoder()
cat_1hot_mtx = encoder.fit_transform(feature_df[cat_feature])
cat_1hot_df = pd.DataFrame(cat_1hot_mtx.todense(),columns=encoder.get_feature_names(),index=feature_df.index) # Convert to dataframe
cat_1hot_df
```

```python
# Append one-hot encoded features
feature_df = pd.concat([feature_df,cat_1hot_mtx],axis=1)
feature_df
```

```python
# Drop original categorical features
feature_df = feature_df.drop(cat_feature, axis=1)
feature_df
```

```python
# Export feature table to file
feature_df.to_csv("../Data/Features/features_{}.csv".format(datetime.datetime.now().strftime("%Y%m%d")))
```
